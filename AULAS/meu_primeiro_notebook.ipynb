```python
# prompt: faca o download do dataset iris em um dataframe df , baixe diretamente da url do UCI

import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
df = pd.read_csv(url, names=names)
```

Estamos lendo uma base que fala sobre a flor 'iris'

```python
# dados das flores iris
print(df)
```

```python
# prompt: para meu df com os dados do iris crie um hotograma para cada uma das features

import matplotlib.pyplot as plt

# Create histograms for each feature
for col in df.columns[:-1]:  # Exclude the 'class' column
  plt.figure()
  df[col].hist()
  plt.title(f'Histogram of {col}')
  plt.xlabel(col)
  plt.ylabel('Frequency')
  plt.show()
```
