```python
# prompt: Agora divida dados X e Y, em X remova a coluna ID, e para Y use a coluna default payment next month

import pandas as pd

# URL da base de dados
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00350/default%20of%20credit%20card%20clients.xls"

# Lê a base de dados usando pandas, pulando a primeira linha (cabeçalho)
dados = pd.read_excel(url, header=1)

# Cria novas colunas com base nos valores categóricos de 'SEX', 'EDUCATION' e 'MARRIAGE'
dados = pd.get_dummies(dados, columns=['SEX', 'EDUCATION', 'MARRIAGE'], prefix=['SEX', 'EDUCATION', 'MARRIAGE'])

# Divide os dados em X e Y
X = dados.drop('ID', axis=1)
X = X.drop('default payment next month', axis = 1)
y = dados['default payment next month']
```

```python
X
```

```python
## Como uma abordagem inicial para tratar desbalanceamanto (não ideal) vamos fazer undersampling.
```

```python
# prompt: faça undersampling para valores 0 em default payment next month para equilibrar minha base de dados em X e Y

from imblearn.under_sampling import RandomUnderSampler

# ... (código anterior) ...

# Inicializa o RandomUnderSampler
rus = RandomUnderSampler(random_state=42)  # Defina uma semente aleatória para reprodutibilidade

# Aplica o undersampling aos dados X e y
X_resampled, y_resampled = rus.fit_resample(X, y)

# Agora X_resampled e y_resampled contêm os dados balanceados
# Verifique a contagem dos valores na coluna target após o undersampling
print(y_resampled.value_counts())

# ... (código posterior) ...
```

```python
y_resampled.value_counts()
```

```python
#sobrescrever os nomes:
X, y = X_resampled, y_resampled
```

## Treinamento

```python
# prompt: faça  modelagem usando kfold cross validaton k=10

from sklearn.model_selection import KFold
from sklearn.tree import  DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np


# Inicializa o KFold com k=10
kf = KFold(n_splits=10, shuffle=True, random_state=42) # Defina uma semente aleatória para reprodutibilidade

# Inicializa a lista para armazenar as pontuações de acurácia
accuracy_scores = []

# Inicializa o modelo (exemplo: Regressão Logística)
model =  DecisionTreeClassifier(random_state=42)

# Loop pelos folds
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Treina o modelo
    model.fit(X_train, y_train)

    # Faz previsões
    y_pred = model.predict(X_test)

    # Calcula a acurácia
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Imprime as pontuações de acurácia para cada fold
print("Acurácia para cada fold:", accuracy_scores)

# Calcula a acurácia média
mean_accuracy = np.mean(accuracy_scores)
print("Acurácia média:", mean_accuracy)
```

Conclusões:

- Treinamos uma Árvore de Decisão
- Notamos que na base de dados temos campos que podem fazer discriminação sexista e etarista. Extremamente importante discutir o uso correto desses dados, com relação a computação justa https://fairlearn.org/
- E para os ajustes da base de dados o resultado inicial não se mostrou tão bom, cabendo novos experimentos. Acurácia média: 0.62 (pior caso é o 0.5 'moeda')
