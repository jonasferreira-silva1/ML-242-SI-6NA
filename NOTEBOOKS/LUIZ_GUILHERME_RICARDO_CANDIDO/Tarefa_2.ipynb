#Análise Exploratória de Dados - Vinhos (UCI)
Importando o dataset e carregando como dataframe no pandas

```python
import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

#Tarefas
#1. Análise Exploratória de Variáveis
**Realize uma análise exploratória básica das seguintes colunas, fornecendo estatísticas descritivas e insights:**

classe: Faça a contagem das diferentes classes presentes no dataset.

*   alcool
*   acido_malico
*   magnesio
*   matiz



```python
classe_count = vinhos['classe'].value_counts()
classe_stats = vinhos['classe'].describe()

display(classe_count)
display(classe_stats)

stats = vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe()

display(stats)
```

#2. Criação de Boxplots
**Crie boxplots para explorar a dispersão dos dados nas seguintes colunas:**
* magnesio
* prolina
* cinzas
* alcool

```python
import matplotlib.pyplot as plt

box = ['magnesio', 'prolina', 'cinzas', 'alcool']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

for x, coluna in enumerate(box):
    linha = x // 2
    colunA = x % 2
    axes[linha, colunA].boxplot(vinhos[coluna])
    axes[linha, colunA].set_title(f'Boxplot {coluna}')

plt.tight_layout()

plt.show()
```

#3. Boxplots Segregados por Classe
**Crie um boxplot das seguintes colunas, segregando os dados pela variável classe:**
* magnesio
* fenois_totais
* acido_malico
* alcool

```python
vinhos.boxplot(column='magnesio', by='classe', figsize=(6, 6))
plt.title('Boxplot - Magnésio por Classe')
plt.suptitle('')
plt.show()

vinhos.boxplot(column='fenois_totais', by='classe', figsize=(6, 6))
plt.title('Boxplot - Fenois Totais por Classe')
plt.suptitle('')
plt.show()

vinhos.boxplot(column='alcool', by='classe', figsize=(6, 6))
plt.title('Boxplot - Álcool por Classe')
plt.suptitle('')
plt.show()

vinhos.boxplot(column='acido_malico', by='classe', figsize=(6, 6))
plt.title('Boxplot - Ácido Málico por Classe')
plt.suptitle('')
plt.show()
```

#4. Criação de Histograma
**Crie um histograma para visualizar a distribuição dos dados na coluna:
fenois_totais, utilizando 10 bins**

```python
plt.figure(figsize=(8, 6))
plt.hist(vinhos['fenois_totais'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma - Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```

```python
classe3 = vinhos[vinhos['classe'] == '3']

plt.figure(figsize=(8, 6))
plt.hist(classe3['fenois_totais'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma - Fenois Totais (Classe 3)')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
