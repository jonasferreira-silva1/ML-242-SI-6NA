# Análise Exploratória de Dados

## Importando as bibliotecas necessárias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

## Configurações para os gráficos

```python
sns.set(style='whitegrid')
```

## URL do dataset Wine

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data' 
```

## Nome das colunas em pt-br

```python
colunas_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]
```

## Ler o arquivo CSV especificando as colunas, definindo a coluna 'classe' como object


```python
vinhos = pd.read_csv(url, names=colunas_pt_br, dtype={'classe': object})
```

# 1. Análise Exploratória de Variáveis

## Estatísticas descritivas para as colunas

```python
print("Estatísticas descritivas:")
vinhos.describe()
```

## Contagem de classes

```python
print("\nContagem das classes:")
print(vinhos['classe'].value_counts())
```

# 2. Criação de Boxplots

## Criando boxplots para explorar a dispersão dos dados

## Boxplot para 'magnesio'

```python
sns.boxplot(data=vinhos, y='magnesio')
plt.title('Boxplot - Magnésio')
```


## Boxplot para 'prolina'

```python
sns.boxplot(data=vinhos, y='prolina')
plt.title('Boxplot - Prolina')
```


## Boxplot para 'cinzas'

```python
sns.boxplot(data=vinhos, y='cinzas')
plt.title('Boxplot - Cinzas')
```

## Boxplot para 'alcool'

```python
sns.boxplot(data=vinhos, y='alcool')
plt.title('Boxplot - Álcool')
plt.tight_layout()
plt.show()
```

# 3. Boxplots Segregados por Classe

## Boxplot segregado por classe para várias colunas

## Boxplot para 'magnesio' por classe

```python
vinhos.boxplot(column='magnesio', by='classe')
plt.title('Boxplot - Magnésio por Classe')
plt.suptitle('')  
```

## Boxplot para 'fenois_totais' por classe

```python
vinhos.boxplot(column='fenois_totais', by='classe')
plt.title('Boxplot - Fenóis Totais por Classe')
plt.suptitle('')  
```

# Boxplot para 'acido_malico' por classe

```python
vinhos.boxplot(column='acido_malico', by='classe')
plt.title('Boxplot - Ácido Málico por Classe')
plt.suptitle('')  
```

## Boxplot para 'alcool' por classe

```python
vinhos.boxplot(column='alcool', by='classe')
plt.title('Boxplot - Álcool por Classe')
plt.suptitle('') 
plt.tight_layout()
plt.show()
```

# 4. Criação de Histograma

## Histograma para a coluna 'fenois_totais' com 10 bins

```python
plt.figure(figsize=(6, 4))
sns.histplot(vinhos['fenois_totais'], bins=10, kde=True, edgecolor='black')
plt.title('Histograma - Fenóis Totais')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()
```

## Histograma para 'fenois_totais', filtrando apenas a classe 3

```python
plt.figure(figsize=(6, 4))
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']
plt.hist(vinhos_classe_3['fenois_totais'], bins=10, edgecolor='black')
plt.title('Histograma - Fenóis Totais (Classe 3)')
plt.xlabel('Fenóis Totais')
plt.ylabel('Frequência')
plt.show()
```
