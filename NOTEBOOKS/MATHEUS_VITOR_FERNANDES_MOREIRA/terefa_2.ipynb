```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe', 'alcool', 'acido_malico', 'cinzas', 'alcalinidade_de_cinzas',
    'magnesio', 'fenois_totais', 'flavanoides', 'fenois_nao_flavanoides',
    'proantocianinas', 'intensidade_de_cor', 'matiz',
    'od280_od315_de_vinhos_diluidos', 'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

# 1. Análise Exploratória de Variáveis
print("Análise Exploratória de Variáveis\n")

# a. Contagem das diferentes classes
print("Contagem das diferentes classes:")
print(vinhos['classe'].value_counts(), "\n")

# b. Estatísticas descritivas
print("Estatísticas descritivas das colunas selecionadas:")
print(vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe(), "\n")

# 2. Criação de Boxplots
plt.figure(figsize=(12, 10))

# Boxplot de Magnésio
plt.subplot(2, 2, 1)
sns.boxplot(x=vinhos['magnesio'])
plt.title('Boxplot de Magnésio')
plt.xlabel('Concentração de Magnésio')

# Boxplot de Prolina
plt.subplot(2, 2, 2)
sns.boxplot(x=vinhos['prolina'])
plt.title('Boxplot de Prolina')
plt.xlabel('Concentração de Prolina')

# Boxplot de Cinzas
plt.subplot(2, 2, 3)
sns.boxplot(x=vinhos['cinzas'])
plt.title('Boxplot de Cinzas')
plt.xlabel('Concentração de Cinzas')

# Boxplot de Álcool
plt.subplot(2, 2, 4)
sns.boxplot(x=vinhos['alcool'])
plt.title('Boxplot de Álcool')
plt.xlabel('Concentração de Álcool')

plt.tight_layout()
plt.show()

# 3. Boxplots Segregados por Classe
plt.figure(figsize=(12, 10))

# Boxplot de Magnésio por Classe
plt.subplot(2, 2, 1)
sns.boxplot(x='classe', y='magnesio', data=vinhos)
plt.title('Boxplot de Magnésio por Classe')
plt.xlabel('Classe')
plt.ylabel('Concentração de Magnésio')

# Boxplot de Fenois Totais por Classe
plt.subplot(2, 2, 2)
sns.boxplot(x='classe', y='fenois_totais', data=vinhos)
plt.title('Boxplot de Fenois Totais por Classe')
plt.xlabel('Classe')
plt.ylabel('Concentração de Fenois Totais')

# Boxplot de Ácido Málico por Classe
plt.subplot(2, 2, 3)
sns.boxplot(x='classe', y='acido_malico', data=vinhos)
plt.title('Boxplot de Ácido Málico por Classe')
plt.xlabel('Classe')
plt.ylabel('Concentração de Ácido Málico')

# Mantendo o último gráfico de Álcool
plt.subplot(2, 2, 4)
sns.boxplot(x='classe', y='alcool', data=vinhos)
plt.title('Boxplot de Álcool por Classe')
plt.xlabel('Classe')
plt.ylabel('Concentração de Álcool')

plt.tight_layout()
plt.suptitle('Boxplots Segregados por Classe', y=1.02)
plt.show()


# 4. Criação de Histogramas
# Histograma de Fenois Totais
plt.figure(figsize=(10, 5))
sns.histplot(vinhos['fenois_totais'], bins=10, kde=True)
plt.title('Histograma de Fenois Totais')
plt.xlabel('Concentração de Fenois Totais')
plt.ylabel('Frequência')
plt.show()

# Histograma de Fenois Totais (Apenas Classe 3)
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']
plt.figure(figsize=(10, 5))
sns.histplot(vinhos_classe_3['fenois_totais'], bins=10, kde=True)
plt.title('Histograma de Fenois Totais (Classe 3)')
plt.xlabel('Concentração de Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
