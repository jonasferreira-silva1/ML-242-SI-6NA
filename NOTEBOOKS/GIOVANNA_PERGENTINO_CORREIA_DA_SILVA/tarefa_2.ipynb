Exercício Prático: Análise Exploratória de Dados - Vinhos (UCI)

Faça a leitura dos dados sobre vinhos da UCI, utilizando o seguinte código:

```python
import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

1. Análise Exploratória de Variáveis
Realize uma análise exploratória básica das seguintes colunas, fornecendo estatísticas descritivas e insights:

```python
contagem_classes = vinhos['classe'].value_counts()
print("Contagem das classes:\n", contagem_classes)
```

```python
descricao_alcool = vinhos['alcool'].describe()
contagem_alcool = vinhos['alcool'].value_counts()
print("\nEstatísticas descritivas da coluna 'alcool':\n", descricao_alcool)
print("Contagem da coluna alcool:\n", contagem_alcool)

print("\nInsights sobre a coluna 'alcool':")
print("- A média de teor alcoólico é de", round(vinhos['alcool'].mean(), 2))
print("- O teor alcoólico varia de", vinhos['alcool'].min(), "a", vinhos['alcool'].max())
```

```python
descricao_acido_malico = vinhos['acido_malico'].describe()
contagem_acido_malico = vinhos['acido_malico'].value_counts()
print("\nEstatísticas descritivas da coluna 'acido_malico':\n", descricao_acido_malico)
print("Contagem da coluna acido_malico:\n", contagem_acido_malico)

print("\nInsights sobre a coluna 'acido_malico':")
print("- A média de ácido málico é de", round(vinhos['acido_malico'].mean(), 2))
print("- O ácido málico varia de", vinhos['acido_malico'].min(), "a", vinhos['acido_malico'].max())
```

```python
descricao_magnesio = vinhos['magnesio'].describe()
contagem_magnesio = vinhos['magnesio'].value_counts()
print("\nEstatísticas descritivas da coluna 'magnesio':\n", descricao_magnesio)
print("Contagem da coluna magnesio:\n", contagem_magnesio)

print("\nInsights sobre a coluna 'magnesio':")
print("- A média de magnésio é de", round(vinhos['magnesio'].mean(), 2))
print("- O magnésio varia de", vinhos['magnesio'].min(), "a", vinhos['magnesio'].max())
```

```python
descricao_matiz = vinhos['matiz'].describe()
contagem_matiz = vinhos['matiz'].value_counts()
print("\nEstatísticas descritivas da coluna 'matiz':\n", descricao_matiz)
print("Contagem da coluna matiz:\n", contagem_matiz)

print("\nInsights sobre a coluna 'matiz':")
print("- A média de matiz é de", round(vinhos['matiz'].mean(), 2))
print("- A matiz varia de", vinhos['matiz'].min(), "a", vinhos['matiz'].max())
```

2. Criação de Boxplots
Crie boxplots para explorar a dispersão dos dados nas seguintes colunas:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.boxplot(vinhos['magnesio'])
plt.title('Boxplot de Magnésio')

plt.subplot(2, 2, 2)
plt.boxplot(vinhos['prolina'])
plt.title('Boxplot de Prolina')

plt.subplot(2, 2, 3)
plt.boxplot(vinhos['cinzas'])
plt.title('Boxplot de Cinzas')

plt.subplot(2, 2, 4)
plt.boxplot(vinhos['alcool'])
plt.title('Boxplot de Álcool')

plt.tight_layout()
plt.show()
```

3. Boxplots Segregados por Classe
Crie um boxplot das seguintes colunas, segregando os dados pela variável classe:

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {'classe': ['Vinho A', 'Vinho A', 'Vinho B', 'Vinho B', 'Vinho C', 'Vinho C'],
        'magnesio': [120, 115, 130, 125, 140, 135],
        'fenois_totais': [100, 110, 120, 130, 140, 150],
        'acido_malico': [5, 6, 7, 8, 9, 10],
        'alcool': [10, 11, 12, 13, 14, 15]}
vinhos = pd.DataFrame(data)

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

plt.ion()

columns = ['magnesio', 'fenois_totais', 'acido_malico', 'alcool']
titles = ['Boxplot de Magnésio por Classe', 'Boxplot de Fenois Totais por Classe',
           'Boxplot de Ácido Málico por Classe', 'Boxplot de Álcool por Classe']

for i, col in enumerate(columns):
    ax = axes.flat[i]
    vinhos.boxplot(column=col, by='classe', ax=ax)
    ax.set_title(titles[i])

plt.tight_layout()
plt.show()
```

4. Criação de Histograma
Crie um histograma para visualizar a distribuição dos dados na coluna:

```python
import matplotlib.pyplot as plt
plt.hist(vinhos['fenois_totais'], bins=10)
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.title('Histograma de Fenois Totais')
plt.show()
```

Crie um histograma da coluna fenois_totais, filtrando apenas a classe 3, utilizando 10 bins.

```python
import matplotlib.pyplot as plt
plt.hist(vinhos[vinhos['classe'] == '3']['fenois_totais'], bins=10)
plt.xlabel('Fenois Totais (Classe 3)')
plt.ylabel('Frequência')
plt.title('Histograma de Fenois Totais (Classe 3)')
plt.show()
```
