# Importando Bibliotecas e DataFrame

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

# Análise Exploratória de Variáveis

```python
vinhos.describe()
```

```python
vinhos['classe'].value_counts()
```

# Criação de Boxplots

```python
sns.boxplot(y=vinhos['magnesio'],)
plt.title('Boxplot - Magnésio')
plt.show()
```

```python
sns.boxplot(y=vinhos['prolina'],)
plt.title('Boxplot - Prolina')
plt.show()
```

```python
sns.boxplot(y=vinhos['cinzas'],)
plt.title('Boxplot - Cinzas')
plt.show()
```

```python
sns.boxplot(y=vinhos['alcool'],)
plt.title('Boxplot - Álcool')
plt.show()
```

# Boxplots Segregados por Classe

```python
vinhos.boxplot(column='magnesio', by='classe', grid=False)
plt.xlabel('Classe')
plt.ylabel('Magnésio')
plt.show()
```

```python
vinhos.boxplot(column='fenois_totais', by='classe', grid=False)
plt.xlabel('Classe')
plt.ylabel('Fenois Totais')
plt.show()
```

```python
vinhos.boxplot(column='acido_malico', by='classe', grid=False)
plt.xlabel('Classe')
plt.ylabel('Ácido Málico')
plt.show()
```

```python
vinhos.boxplot(column='alcool', by='classe', grid=False)
plt.xlabel('Classe')
plt.ylabel('Álcool')
plt.show()
```

# Criação de Histograma

```python
sns.histplot(vinhos['fenois_totais'], bins=10, kde=False,)
plt.title('Histograma - Fenois com 10 bins')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```

```python
vinhos_3 = vinhos[vinhos['classe'] == '3']
sns.histplot(vinhos_3['fenois_totais'], bins=10, kde=False,)
plt.title('Histograma - Fenois')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
