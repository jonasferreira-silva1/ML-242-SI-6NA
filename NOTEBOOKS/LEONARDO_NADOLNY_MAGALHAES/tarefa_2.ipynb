# Tarefa 2: Análise Exploratória de Dados - Vinhos (UCI)

Importando o dataset e carregando como dataframe no pandas

```python

import pandas as pd

# URL do dataset Wine
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

# Nome das colunas em português BR
column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

# Ler o arquivo CSV com as colunas especificadas, definindo a coluna 'classe' como object
vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

## 1. Análise Exploratória de Variáveis

### Realize uma análise exploratória básica das seguintes colunas, fornecendo estatísticas descritivas e insights:


- classe: Faça a contagem das diferentes classes presentes no dataset.
- alcool
- acido_malico
- magnesio
- matiz


```python

classe_count = vinhos['classe'].value_counts()
classe_stats = vinhos['classe'].describe()

print("Estatísticas da coluna 'classe':")
display(classe_count)
display(classe_stats)

stats = vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe()

print("\nEstatísticas das colunas numéricas:")
display(stats)
```

## 2. Criação de Boxplots

### Crie boxplots para explorar a dispersão dos dados nas seguintes colunas:

- magnesio
- prolina
- cinzas
- alcool

```python
import matplotlib.pyplot as plt
```

```python
col_box = ['magnesio', 'prolina', 'cinzas', 'alcool']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

for i, coluna in enumerate(col_box):
    row = i // 2  
    col = i % 2   
    axes[row, col].boxplot(vinhos[coluna])  
    axes[row, col].set_title(f'Boxplot de {coluna}')  

plt.tight_layout()

plt.show()
```

## 3. Boxplots Segregados por Classe

### Crie um boxplot das seguintes colunas, segregando os dados pela variável classe:


- magnesio
- fenois_totais
- acido_malico
- alcool

```python
coluna_box = ['magnesio', 'fenois_totais', 'acido_malico', 'alcool']

for coluna in coluna_box:
    vinhos.boxplot(column=coluna, by='classe', figsize=(8, 6))
    plt.title(f'Boxplot de {coluna} segregado por Classe')
    plt.suptitle('')  
    plt.xlabel('Classe')
    plt.ylabel(coluna)
    plt.show()
```

## 4. Criação de Histograma

### Crie um histograma para visualizar a distribuição dos dados na coluna:

fenois_totais, utilizando 10 bins

Dica: Utilize plt.hist() ou sns.histplot() para criar histogramas e ajuste o número de bins usando o argumento bins=10.

```python
plt.figure(figsize=(8, 6))
plt.hist(vinhos['fenois_totais'], bins=10, edgecolor='black')

plt.title('Histograma de Fenois Totais')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')

plt.show()
```

### Crie um histograma da coluna fenois_totais, filtrando apenas a classe 3, utilizando 10 bins.

Dica: Filtre os dados com vinhos[vinhos['classe'] == '3'] antes de gerar o histograma.

```python
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']

plt.figure(figsize=(8, 6))
plt.hist(vinhos_classe_3['fenois_totais'], bins=10, edgecolor='black')

plt.title('Histograma de Fenois Totais - Classe 3')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')

plt.show()
```
