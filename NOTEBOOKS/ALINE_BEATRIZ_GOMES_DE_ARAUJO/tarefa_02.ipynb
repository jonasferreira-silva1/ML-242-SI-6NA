```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]

vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})
```

```python
# Tarefa 1: Análise Exploratória de Variáveis
# Contagem das classes
print("Contagem das Classes:")
print(vinhos['classe'].value_counts())

# Estatísticas descritivas das colunas solicitadas
print("\nEstatísticas descritivas - Álcool:")
print(vinhos['alcool'].describe())

print("\nEstatísticas descritivas - Ácido Málico:")
print(vinhos['acido_malico'].describe())

print("\nEstatísticas descritivas - Magnésio:")
print(vinhos['magnesio'].describe())

print("\nEstatísticas descritivas - Matiz:")
print(vinhos['matiz'].describe())
```

```python

# Tarefa 2: Criação de Boxplots
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.boxplot(data=vinhos, y='magnesio')
plt.title("Boxplot - Magnésio")

plt.subplot(2, 2, 2)
sns.boxplot(data=vinhos, y='prolina')
plt.title("Boxplot - Prolina")

plt.subplot(2, 2, 3)
sns.boxplot(data=vinhos, y='cinzas')
plt.title("Boxplot - Cinzas")

plt.subplot(2, 2, 4)
sns.boxplot(data=vinhos, y='alcool')
plt.title("Boxplot - Álcool")

plt.tight_layout()
plt.show()
```

```python
# Tarefa 3: Boxplots Segregados por Classe
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.boxplot(data=vinhos, x='classe', y='magnesio')
plt.title("Magnésio por Classe")

plt.subplot(2, 2, 2)
sns.boxplot(data=vinhos, x='classe', y='fenois_totais')
plt.title("Fenois Totais por Classe")

plt.subplot(2, 2, 3)
sns.boxplot(data=vinhos, x='classe', y='acido_malico')
plt.title("Ácido Málico por Classe")

plt.subplot(2, 2, 4)
sns.boxplot(data=vinhos, x='classe', y='alcool')
plt.title("Álcool por Classe")

plt.tight_layout()
plt.show()
```

```python
# Tarefa 4: Criação de Histograma
# Histograma para fenois_totais com 10 bins
plt.figure(figsize=(10, 6))
sns.histplot(vinhos['fenois_totais'], bins=10)
plt.title('Histograma - Fenois Totais (10 bins)')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()

# Histograma para fenois_totais apenas da classe 3
vinhos_classe_3 = vinhos[vinhos['classe'] == '3']
plt.figure(figsize=(10, 6))
sns.histplot(vinhos_classe_3['fenois_totais'], bins=10)
plt.title('Histograma - Fenois Totais (Classe 3)')
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.show()
```
